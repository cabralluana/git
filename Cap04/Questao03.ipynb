```python
import pandas as pd

# É necessário mudar o caminho em outro computador
covid19CleanComplete = pd.read_csv("Arquivos csv/covid_19_clean_complete.csv")
```

```python
# Letra a - Elabore uma função que tenha uma linha do banco de dados como 
# argumento e, se a coluna "Province/State" não for valor faltante, concatenar
# seu valor com coluna "Country/Region"
def verificarProvinceState(linha):
    """"É uma função que vai receber uma linha do dataset e verificar se ela é
    NaN e se não for irá concatenar a coluna 'Province/State' com 'Country/Region'"""
    if pd.notna(linha['Province/State']) == True:
        linha['Country/Region'] = linha['Country/Region'] + "_" + linha[
            'Province/State']
        return linha
    return linha
```

```python
# Letra b - Faça uma cópia do banco de dados para poder retirar informações sem
# perder o banco de dados original
covid19CleanCompleteCopia = covid19CleanComplete
print(covid19CleanCompleteCopia)
```

```python
# Letra c - Aplique, por meio da método "apply", a função criada no item a
covid19CleanCompleteCopia = covid19CleanCompleteCopia.apply(verificarProvinceState, axis=1)
print(covid19CleanCompleteCopia)
```

```python
# Letra d - Exclua a coluna "Province/State" do novo dataframe
covid19CleanCompleteCopia = covid19CleanCompleteCopia.drop(["Province/State"], axis = 1)
print(covid19CleanCompleteCopia)
```
